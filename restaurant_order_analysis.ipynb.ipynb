```python
import plotly.express as px
import pandas as pd

# 1. Definindo os novos dados (cálculos já aplicados)
# Local 1: 18.650 (estável)
# Local 2: 15.600 (aumento de 30%)
# Local 3: 10.300 (aumento de 3%)
# Local 4: 15.100 (estável)
# Restaurante Principal: 10.000 (queda de 50%)

dados_atualizados = {
    'Unidade': ['Local 1', 'Local 2', 'Local 3', 'Local 4', 'Restaurante Principal'],
    'Pedidos': [18650, 15600, 10300, 15100, 10000]
}

# Criando o DataFrame
df = pd.DataFrame(dados_atualizados)

# 2. Criando o gráfico de pizza
# Usei 'hole=0.3' para um visual de rosca, se preferir pizza total, mude para 0
fig = px.pie(df, 
             values='Pedidos', 
             names='Unidade', 
             title='Análise de Pedidos: Impacto das Reformas e Inspeções',
             hole=0.3)

# 3. Colocando nome e porcentagem dentro das fatias
fig.update_traces(textinfo='label+percent', textposition='inside')

# Exibindo o resultado
fig.show()
```
